```python
# function to print summary dataframe in colunm type
def print_data_summary(df):
  # count the number of categorial colunms
  object_count=df.select_dtypes(include=['object']).shape[1]
  # count the number of float item
  float_count=df.select_dtypes(include=['float']).shape[1]
  # count the number of integer column
  int_count=df.select_dtypes(include=['int']).shape[1]

  # print the count in a bullet point
  print(f".object_columns :{object_count}")
  print(f".Float_columns : {float_count}")
  print(f".int_columns : {int_count}")

# print the heading of the data section
print_data_summary(df)
```

BAR PLOT

```python
# create the sample data
data={
    'region':['Andhra Pradesh','Tripura','Meghalaya','Uttar Pradesh','Jammu & Kasmir'],
    'date': ['31-05-2019', '30-06-2019', '31-07-2019', '31-08-2019', '30-09-2019'],
    'estimated_unemployment_rate': [3.65, 3.05, 3.75, 3.32, 5.17],
    'area': ['Rural', 'Urban', 'Rural', 'Urban', 'Rural']
}

# create the dataframe from a sample data
df_1=pd.DataFrame(data)

# convert the date column to datetime format
df_1['Date']=pd.to_datetime(df_1['date'],format='%d-%m-%Y')


# create the line plot using plolty
fig=px.bar(df_1,
           x='date',
        y='estimated_unemployment_rate',
        color='region',
        title='Estimated Unemployment Rate Over Time by Region',
        labels={
            'date': 'Date',
            'estimated_unemployment_rate': 'Estimated Unemployment Rate (%)',
            'region': 'Region',
            'area' : 'Area'
        },
        barmode='stack'
    )

# # show the plot
fig.show()
```


SCATTER PLOT

```python
fig=px.scatter(df,x=' Estimated Labour Participation Rate (%)',
               y=' Estimated Unemployment Rate (%)',
               color='Region',
               size=' Estimated Employed',
               hover_data=['Area'],
               title='Relationship between Estimated Labor Participation Rate and Estimated Unemployment Rate')
fig.show()
```

```python
# create the heatmap using plotly express
fig = px.density_heatmap(
    df,
    x=' Date',
    y='Region',
    z=' Estimated Labour Participation Rate (%)',
    histfunc='avg',
    title='Estimated Labour Participation Heatmap',
    labels={
        ' Date': 'Date',
        'Region': 'Region',
        ' Estimated Labour Participation Rate (%)': 'Estimated Labour Participation Rate (%)'
    },
    color_continuous_scale='Viridis'  # Color intensity scale
)

# Add colorbar title
fig.update_layout(coloraxis_colorbar=dict(title='Labour Participation Rate(%)'))

# Show the interactive plot
fig.show()
```

```python
fig = px.bar(df,x='Region',y=' Estimated Unemployment Rate (%)',animation_frame='Month_name',color='Region',
            title='Unemployment rate from Jan 2020 to Oct 2020(State)',color_continuous_scale='RdBu_r'
            )

fig.update_layout(xaxis={'categoryorder':'total descending'})
fig.show()
```
