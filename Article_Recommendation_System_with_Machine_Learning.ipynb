```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction import text
from sklearn.metrics.pairwise import cosine_similarity
data=pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/articles.csv", encoding='latin1')

data.head()
```

```python
article=data["Article"].tolist()
uni_tfidf=text.TfidfVectorizer(stop_words="english")
uni_metrix=uni_tfidf.fit_transform(article)
uni_sim=cosine_similarity(uni_metrix)
def recommand_article(x):
    return",".join(data["Title"].loc[x.argsort()[-5:-1]])
data["Recommanded Article"]=[recommand_article(x) for x in uni_sim]
data.head()
```
