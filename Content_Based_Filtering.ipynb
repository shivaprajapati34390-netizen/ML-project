```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import linear_kernel
data = pd.read_csv('https://query.data.world/s/mran7gzcmlwbbbxngv6nomplkkuijj?dws=00000')
data.head()
```

```python
tfid=TfidfVectorizer(stop_words="english")
data['description']=data['description'].fillna('')
overview_matrix=tfid.fit_transform(data['description'])
similarity_matrix=linear_kernel(overview_matrix)
mapping=pd.Series(data.index,index=data['name'])
print(mapping)
```

```python
# lets create the function and have a look at how recommand system working
```

```python
def recommand_movies(movie_name):
  movie_index = mapping[movie_name]
  similarity_score = list(enumerate(similarity_matrix[movie_index]))
  similarity_score = sorted(similarity_score, key=lambda x:x[1], reverse=True)
  similarity_score = similarity_score[1:15] # Get top 14 similar movies (excluding itself)

  movie_indices = [i[0] for i in similarity_score]
  return data['name'].iloc[movie_indices]

print(recommand_movies('The Haunting of Hill House')) # Using an existing movie name for demonstration
```
